```python
# Netflix Titles Dataset Summary

# This dataset contains 8,807 records and 12 features describing Movies and TV Shows available on Netflix.
# It includes information such as title, director, cast, country, release year, rating, duration, and genre.
# The dataset contains some missing values in columns like director, cast, country, and rating.
# duration,cast and decription would not be useful during anaylysis ,so we will drop them.
# date_added column need to be casted into date time.
# Most features are categorical, while release_year is the main numerical column.
# listed_in has come seperated values and white spaces that needed to be handle and casted into category.
```

```python
df.head()
```

```python
df['date_added']= (df['date_added'].str.strip())
df['date_added']= pd.to_datetime(df['date_added'],errors='coerce')
print(df['date_added'].dtypes)
```

```python
df['date_added'] = df['date_added'].dt.year
df['date_added'].dtypes
```

```python
df.head()
```

```python
# dropping the columns
cols = ['description',"cast",'duration']
df.drop(cols,axis=1,inplace=True)
df.columns
```

```python
df.head()
```

Q1.How has Netflix content grown over the years?

```python
yearly_growth = df['date_added'].value_counts().sort_index()
yearly_growth.plot(kind='line',marker='*',figsize = (8,5))
plt.title('Netflix Content Growth Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

1.Netflix content addition was relatively slow before 2015.
2.A sharp increase is observed after 2015.
3.Peak content addition occurred around 20182019.
4.Slight decline after 2020 may reflect content saturation or data cutoff.

Q2.which type(Movies or TV shows) dominates the plateform?

```python
sns.countplot(data=df,x='type',hue='type',palette='pastel',legend=True)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

1.Movies significantly outnumber TV Shows.
2.TV Shows represent a smaller but still substantial portion of the catalog.

Q3.which countries produce the most content?

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh',color="skyblue")
plt.xlabel('Number of Content')
plt.ylabel('Country')
plt.title('Top 10 Countries by Netflix Content')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

1.The United States dominates content production.
2.India and the United Kingdom follow.
3.Strong presence of international countries indicates global diversity.

Q4.what are the most common rating categories?

```python
# claening data 
valid_ratings = ['TV-MA','TV-14','TV-PG','R','PG-13','PG','G','NR','TV-Y','TV-Y7','TV-G','TV-Y7-FV','NC-17','UR']
df = df[df['rating'].isin(valid_ratings)]

sns.countplot(data=df,y='rating',order=df['rating'].value_counts().index,hue = 'rating',palette='coolwarm',legend=False)
plt.title('Netflix Content By Rating')
plt.tight_layout()
plt.show()
```

1.TV-MA is the most common rating.
2.TV-14 and R-rated content also appear frequently.
3.Family-friendly ratings are comparatively lower.

Q5.what geners are the most popular?

```python
geners = df['listed_in'].str.split(',').explode()
top_geners = geners.value_counts().head(10)
sns.barplot(x=top_geners.values,y=top_geners.index,palette='viridis',hue=top_geners.index)
plt.title('Top 10 Netflix Genres')
plt.ylabel('Genres')
plt.xlabel('Number of Titles')
plt.tight_layout()
plt.show()
```

1.Drama is the most dominant genre.
2.International Movies and Comedies are also highly represented.
3.Wide variety of genres suggests diversified catalog strategy.
